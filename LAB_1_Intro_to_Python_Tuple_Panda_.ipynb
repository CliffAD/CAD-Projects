```python
print("Problem 1: Confusion Matrix Metrics (Python as a Calculator)")
```

```python
# True Positives represent the number of flood cases correctly predicted as flood
TP = 50

# False Positives represent the number of non flood cases incorrectly predicted as flood
FP = 8

# True Negatives represent the number of non flood cases correctly predicted as no flood
TN = 32

# False Negatives represent the number of flood cases incorrectly predicted as no flood
FN = 10

# Calculate the total number of predictions
total_predictions = TP + FP + TN + FN

# Calculate accuracy using the given formula
accuracy = (TP + TN) / (TP + FP + TN + FN)

# Calculate precision
precision = TP / (TP + FP)

# Calculate recall
recall = TP / (TP + FN)

# Calculate the F1 score
f1_score = (2 * precision * recall) / (precision + recall)

# result part
print("Calculated Results:")

# Print the total number of predictions
print("Total Predictions:", total_predictions)

# Print the accuracy
print("Accuracy:", accuracy)

# Print the precision
print("Precision:", precision)

# Print the recall
print("Recall:", recall)

# Print the F1 score
print("F1 Score:", f1_score)

```

```python
print("Problem 2: R² (Coefficient of Determination)")
```

```python
# Sum of Squares due to Regression represents the explained variation
SSR = 420

# Sum of Squares due to Error represents the unexplained variation
SSE = 180

# Calculate the Total Sum of Squares by adding SSR and SSE
SST = SSR + SSE

# Calculate R-squared using the given formula
r_squared = SSR / SST

# Print the calculated results
print("Calculated Results:")

# Print the Total Sum of Squares
print("Total Sum of Squares (SST):", SST)

# Print the R-squared value
print("R-squared (R²):", r_squared)

# Interpret whether the model explains more than 70% variance
if r_squared > 0.7:
    print("Interpretation: The model explains more than 70% of the variance.")
else:
    print("Interpretation: The model explains less than or equal to 70% of the variance.")
```

```python
print("Problem 3: Mean Squared Error (MSE) Using Python as Calculator")
```

```python
# Store the given squared errors for each observation
e1 = 2
e2 = 4
e3 = 1
e4 = 9
e5 = 4

# Calculate the total sum of squared errors
sum_squared_errors = e1 + e2 + e3 + e4 + e5

# Store the number of observations
n = 5

# Calculate Mean Squared Error using the formula MSE = sum of squared errors / number of observations
mse = sum_squared_errors / n

# Print the calculated result
print("Calculated Results:")

# Print the Mean Squared Error
print("Mean Squared Error (MSE):", mse)
```

```python
print("Problem 4: Cook’s Distance (Influential Observation)")
```

```python
# Squared residual value (given)
e_squared = 9

# Mean Squared Error (given)
MSE = 3

# Number of predictors (given)
p = 3

# Leverage value (given)
h_ii = 0.2

# Sample size (given)
n = 25

# Calculate Cook's Distance using the given formula
cooks_distance = (e_squared / (p * MSE)) * (h_ii / ((1 - h_ii) ** 2))

# Calculate the threshold value 4/n
threshold = 4 / n

# Print the calculated results
print("Calculated Results:")

# Print Cook's Distance
print("Cook's Distance:", cooks_distance)

# Print the threshold value
print("Threshold (4/n):", threshold)

# Decide whether the observation is influential
if cooks_distance > threshold:
    print("Interpretation: The observation is influential.")
else:
    print("Interpretation: The observation is not influential.")
```

```python
# Given slope of the regression line
m = 2.5

# Given intercept of the regression line
c = 10

# Given rainfall value
x = 12

# Calculate the predicted value of y using the formula y = mx + c
y = (m * x) + c

# Print the predicted output
print("Calculated Results:")

# Print the predicted value of y
print("Predicted value of y:", y)

# Explain the meaning of slope and intercept
print("Interpretation:")
print("Slope (m) represents the change in crop yield for each unit increase in rainfall.")
print("Intercept (c) represents the crop yield when rainfall is zero.")
```
